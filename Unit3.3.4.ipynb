```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fdistvar = np.random.f(1,10, 100)
```

```python
fmean = np.mean(fdistvar)
fstd = np.std(fdistvar)

plt.hist(fdistvar)

plt.axvline(x=fmean,color='black')
plt.axvline(x=fmean+fstd, color='red')
plt.axvline(x=fmean-fstd, color='red')
plt.show()
```

The f distribution is highly skewed.  The mean works but it is so long tailed you cannot look at the one standard deviation bands.

```python
tdistvar = np.random.standard_t(10, 100)
```

```python
tmean = np.mean(tdistvar)
tstd = np.std(tdistvar)

plt.hist(tdistvar)

plt.axvline(x=tmean,color='black')
plt.axvline(x=tmean+tstd, color='red')
plt.axvline(x=tmean-tstd, color='red')
plt.show()
```

The standard t distribution looks long tailed and skewed with only 100 samples. The mean and standard deviation 
appear to work fine.


```python
gdistvar = np.random.gumbel(0,.1, 100)
```

```python
gmean = np.mean(gdistvar)
gstd = np.std(gdistvar)

plt.hist(gdistvar)

plt.axvline(x=gmean,color='black')
plt.axvline(x=gmean+gstd, color='red')
plt.axvline(x=gmean-gstd, color='red')
plt.show()
```

The gumbel distribution has some large tails and is skewed to the right.  The standard deviation is probably not the best estimate of its dispersion but the mean seems ok.

```python
ldistvar = np.random.laplace(0,1, 100)

lmean = np.mean(ldistvar)
lstd = np.std(ldistvar)

plt.hist(ldistvar)

plt.axvline(x=lmean,color='black')
plt.axvline(x=lmean+lstd, color='red')
plt.axvline(x=lmean-lstd, color='red')
plt.show()
```

The laplace is highly peaked but symmetric. Our measures are probably reasonable.

```python
udistvar = np.random.uniform(0,1, 100)

umean = np.mean(udistvar)
ustd = np.std(udistvar)

plt.hist(udistvar)

plt.axvline(x=umean,color='black')
plt.axvline(x=umean+ustd, color='red')
plt.axvline(x=umean-ustd, color='red')
plt.show()
```

The uniform distribution is only found between 0 and 1 for this example. As the name describes it is uniformly 
distributed.  The mean is useful and the standard deviation appears to mark out about 60% of the distribution.

```python
rdistvar = np.random.rayleigh(5, 100)

rmean = np.mean(rdistvar)
rstd = np.std(rdistvar)

plt.hist(rdistvar)

plt.axvline(x=rmean,color='black')
plt.axvline(x=rmean+rstd, color='red')
plt.axvline(x=rmean-rstd, color='red')
plt.show()
```

The Rayleigh appears skewed.  The mean appears reasonable as central tendency but it is not clear if the standard deviation can be relied upon.

```python
norm1 = np.random.normal(5, .5 , 100)
norm2 = np.random.normal(10, 1 , 100)
norm3 = norm1 + norm2

nmean = np.mean(norm3)
nstd = np.std(norm3)

plt.hist(norm3)
plt.axvline(x=nmean,color='black')
plt.axvline(x=nmean+nstd, color='red')
plt.axvline(x=nmean-nstd, color='red')
plt.show()
```

We can see some skew developing as we added mean 10 higher variance rv's to the mean 5 rv's.  Histogram appears
slightly skewed to the right.
